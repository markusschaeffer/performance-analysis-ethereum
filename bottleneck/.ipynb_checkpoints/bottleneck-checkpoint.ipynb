# Helper Methods

```python
import numpy as np

# outlier removal method: IQR
# returns list without outliers
# http://colingorrie.github.io/outlier-detection.html
# https://machinelearningmastery.com/how-to-use-statistics-to-identify-outliers-in-data/
def remove_outliers_iqr(data):
    quartile_1, quartile_3 = np.percentile(data, [25, 75])
    iqr = quartile_3 - quartile_1
    cut_off = (iqr * 1.5)
    lower_bound = quartile_1 - cut_off
    upper_bound = quartile_3 + cut_off
    # identify outliers
    outliers = [x for x in data if x < lower_bound or x > upper_bound]
    print('Identified outliers: %d' % len(outliers))
    # remove outliers
    outliers_removed = [x for x in data if x >= lower_bound and x <= upper_bound]
    print('Non-outlier observations: %d' % len(outliers_removed))
    return outliers_removed

# annotates plt with y values
def plot_annotate(plt,xvalues,yvalues):
    for i,j in zip(xvalues, yvalues):
        plt.annotate(str(np.round(j,1)),xy=(i-0.15,j))
    
def autolabel_barchart(rects,axis):
    """
    Attach a text label above each bar displaying its height
    """
    for rect in rects:
        height = rect.get_height()
        axis.text(rect.get_x() + rect.get_width()/2., 0.90*height,'%.1f' % round(float(height),1),ha='center', va='bottom')
```

```python
from pymongo import MongoClient
client = MongoClient()

db = client['test']
collection = db['benchmarklogs']
```

# PyMongo Query ->Pandas DataFrame

```python
import numpy as np
import pandas as pd

query = {"$and":[{"$or":[{"approach":1},{"approach":3}]},{"instanceType":"c5.4xlarge"}, {"nodes":1}, {"usedGenesisJson":"genesis_poa_max_performance.json"}, {"maxTransactions": 1000}, {"scenario":"account"}]}

# Make a query to the specific DB and Collection
cursor = collection.find(query)

# Expand the cursor and construct the DataFrame
df = pd.DataFrame(list(cursor))

sample_size = len(df.index)
print("Number of rows:" + str(sample_size))
```

# Pandas DataFrame -> Plot
## Throughput and Latency

```python
import matplotlib.pyplot as plt

data_tps = df[['txPerSecond','approach']]
data_latency = df[['averageDelay','approach']]

title_tps = "Throughput against Transaction Sending Approach \n (1 Node, PoA max performance, n=" + str(sample_size) + ")"
title_latency = "Latency against Transaction Sending Approach \n (1 Node, PoA max performance, n=" + str(sample_size) + ")"

ylabel_tps = "Transactions Per Second (TPS)"
ylabel_latency = "Transaction Latency [seconds]"

#objects = list(['TPS', 'Latency'])

data1_tps = data_tps.loc[df['approach'] == 1].iloc[:,0]
data3_tps = data_tps.loc[df['approach'] == 3].iloc[:,0]

data1_latency = data_latency.loc[df['approach'] == 1].iloc[:,0]
data3_latency = data_latency.loc[df['approach'] == 3].iloc[:,0]

data_tps = [data1_tps, data3_tps]
data_latency = [data1_latency, data3_latency]

print("TPS approach 1 mean: " + str(data1_tps.mean()))
print("TPS approach 1 median: " + str(data1_tps.median()))
print("Latency approach 1 mean: " + str(data1_latency.mean()))
print("Latency approach 1 median: " + str(data1_latency.median()))
print("\n")
print("TPS approach 3 mean: " + str(data3_tps.mean()))
print("TPS approach 3 median: " + str(data3_tps.median()))
print("Latency approach 3 mean: " + str(data3_latency.mean()))
print("Latency approach 3 median: " + str(data3_latency.median()))
print("\n")
```

```python
color_latency="#ff7f0e"
color_tps="#1f77b4"

fig = plt.figure(1,figsize=(6, 8))

# Create an axes instance
ax1 = fig.add_subplot(211)
ax1.set_title(title_tps)
ax1.set_ylabel(ylabel_tps)
# Create the boxplot
bp1 = ax1.boxplot(data_tps, labels=["Approach 1","Approach 2"], patch_artist=True)

plt.figtext(0.38, 0.70, 'Med: ' + str(round(data1_tps.median(),2)), color='black')
plt.figtext(0.38, 0.68, 'Mean: ' + str(round(data1_tps.mean(),2)), color='black')

plt.figtext(0.8, 0.65, 'Med: ' + str(round(data3_tps.median(),2)), color='black')
plt.figtext(0.8, 0.63, 'Mean: ' + str(round(data3_tps.mean(),2)), color='black')


ax2 = fig.add_subplot(212)
ax2.set_title(title_latency)
ax2.set_ylabel(ylabel_tps)
bp2 = ax2.boxplot(data_latency, labels=["Approach 1","Approach 2"], patch_artist=True)
ax2.set_ylabel("Transaction Latency [seconds]")

plt.figtext(0.38, 0.19, 'Med: ' + str(round(data1_latency.median(),2)), color='black')
plt.figtext(0.38, 0.17, 'Mean: ' + str(round(data1_latency.mean(),2)), color='black')

plt.figtext(0.8, 0.13, 'Med: ' + str(round(data3_latency.median(),2)), color='black')
plt.figtext(0.8, 0.11, 'Mean: ' + str(round(data3_latency.mean(),2)), color='black')


## change outline color, fill color and linewidth of the boxes
for box in bp1['boxes']:
    # change outline color
    box.set(color=color_tps, linewidth = 2)
    # change fill color
    box.set(facecolor="w")
    
## change outline color, fill color and linewidth of the boxes
for box in bp2['boxes']:
    # change outline color
    box.set(color=color_latency, linewidth = 2)
    # change fill color
    box.set(facecolor="w")
    
## change color and linewidth of the medians
for median in bp1['medians']:
    median.set(color="k", linewidth=2)
    
## change color and linewidth of the medians
for median in bp2['medians']:
    median.set(color="k", linewidth=2)
    
fig.tight_layout()
fig.savefig("result_bottleneck_poa_throughput_combined_boxplot.svg", format="svg")
```
